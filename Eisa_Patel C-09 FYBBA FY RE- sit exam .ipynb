```python
athlete_events_df['is_medal'] = athlete_events_df['Medal'].notnull().astype(int)

(
    athlete_events_df
    .groupby(['ID', 'Name'])
    ['is_medal']
    .sum()
    .sort_values(ascending = False)
    .head(10)
    .reset_index()
    .rename(columns = {'is_medal': 'Number of Medals'})
)
```
